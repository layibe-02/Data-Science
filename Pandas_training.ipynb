### Fonction iloc

La fonction iloc est utilisée pour accéder et extraire des données à partir d'un DataFrame en fonction de leur position indexée. "iloc" signifie "index location".

dataframe.iloc[row_index, column_index]


```python
# Accès à la première ligne, deuxième colonne
print(df.iloc[0,1])
```

```python
# Accès aux trois premières lignes, première colonne
print(df.iloc[0:3, 0])
```

```python
# Accès à la dernière ligne, dernière colonne
print(df.iloc[-1, -1])
```

```python
# Sélectionner la première ligne 
print(df.iloc[0])
```

```python
# Sélectionner la dernière ligne 
print(df.iloc[-1])
```

#### Définir des listes de variables catégorielles et quantitatives à utiliser dans l'analyse

```python
categorial_var = ["SEXE", "EDUCATION", "MARIAGE", "Payment"]
quantitative_var = ["LIMIT_BAL", "AGE", "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6", "PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT6"]
```

#### Définir des dictionnaires de remplacement pour modifier les valeurs des variables catégorielles dans le DataFrame.

```python
labels = {
    'SEXE': {1 : 'homme', 2 : 'Femme'},
    'EDUCATION' : {1 : 'École supérieure', 2 : 'Université', 3 : 'Lycée', 4 : 'Autres', 5 : 'Autres', 6 : 'Autres', 0 : 'Autres'},
    'MARIAGE' : {1 : 'Marié', 2 : 'Célibataire', 3 : 'Autres', 0 : 'Autres'},
    'Payment' : {0 : 'Pas de défaut', 1 : 'Défaut'}
}
```

#### Copier le DataFrame df, remplacent les valeurs selon le dictionnaire labels et affichent le DataFrame modifié

```python
df_copy = df.copy()
df_copy.replace(labels, inplace=True)
df_copy
```

#### Afficher uniquement les colonnes catégorielles du DataFrame df_copy
